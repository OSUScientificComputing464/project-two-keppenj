```python
import sklearn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#import sklearn_pandas

from sklearn.datasets import fetch_covtype
from pandas import DataFrame

from sklearn.model_selection import train_test_split
from numpy import array
#from sklearn_pandas import DataFrameMapper

from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler

forest_data = sklearn.datasets.fetch_covtype()

df = pd.DataFrame(data = forest_data.data[:,:])#, columns = forest_data.feature_names())

#print(df)
                  
df_target = pd.DataFrame(data = forest_data.target[:,], columns = ["Cover_type"])

#print(df_target)

models = ["Elevation", "Aspect","Slope", 'Horizontal_Distance_To_Hydrology', \
          'Vertical_Distance_To_Hydrology', 'Horizontal_Distance_To_Roadways',\
         'Hillshade_9am', 'Hillshade_Noon', 'Hillshade_3pm', \
          'Horizontal_Distance_To_Fire_Points', 'Rawah Wilderness', 'Neota Wilderness',\
         'Comanche Peak Wilderness', 'Cache la Poudre Wilderness']

Cover_Type = forest_data.target[:,]

print('Soil Types were left off due to the legnth of time it would take to define')
print('all of them, and the forest_data.feature_name function is not calling them.')

i = 0
for model in models:

    model = forest_data.data[:,i]
    model_title = models[i]
    plt.scatter(model, Cover_Type,  label = model)
    plt.xlabel(model_title)
    plt.ylabel("Cover_Type")
    plt.show()

    i = i+1

```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC

from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler

from numpy import array

X_train, X_test, y_train, y_test = train_test_split(df, df_target)

X_train = array(X_train)
#X_train = X_train.reshape((X_train.shape[0],1))
    
X_test = array(X_test)
#X_test = X_test.reshape((X_test.shape[0],1))
    
y_train = array(y_train)
y_test = array(y_test)
    

logreg = LogisticRegression().fit(X_train, y_train)
logreg_score = logreg.score(X_train, y_train)
logreg_cr_scorei = cross_val_score(logreg, X_train, y_train)
   
logreg_test_cr_scorei = cross_val_score(logreg, X_test, y_test)
logreg_test_cr_score = np.mean(logreg_test_cr_scorei)

print('The average trained cross_val score for Logistic Regression is: ', logreg_cr_scorei)
```
